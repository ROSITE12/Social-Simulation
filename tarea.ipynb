## Ejercicio 1: visualización de puntajes tras varias rondas

```python
# Estrategias
strategies=['Rock','Paper','Scissors']
```

```python
# reglas
payoff={('Rock','Paper'):(0,1),
        ('Paper','Rock'):(1,0),
        ('Rock','Scissors'):(1,0),
        ('Scissors','Rock'):(0,1),
        ('Paper','Scissors'):(0,1),
        ('Scissors','Paper'):(1,0),
        ('Rock','Rock'):(0,0),
        ('Paper','Paper'):(0,0),
        ('Scissors','Scissors'):(0,0)}
```

```python
# nonmbre de los jugadores
names=['Jim','Jane','Peter','Zoe']
```

```python
from random import choice

# Estrategia de juego escogida al azar
choice(strategies);
```

```python
# setting up players
society=[{'name':n,'score':0,'strategy':None} for n in names]
```

```python
# cada jugador tiene un diccionario
society
```

```python
import itertools

# cada par es una dupla de diccionarios
for pair in itertools.combinations(society,2):
    print(pair)
```

```python
import itertools

for player1,player2 in itertools.combinations(society,2):
    print(player1,player2)
```

```python
# restablecer
society=[{'name':n,'score':0,'strategy':None} for n in names]

# varias rondas
for aRound in range(100):

    # en cada ronda:
    for player1,player2 in itertools.combinations(society,2):
        # escogr la estrategia
        player1['strategy']=choice(strategies)

        player2['strategy']=choice(strategies)

        # resultado de cada estartegia escogida
        result=payoff[player1['strategy'],player2['strategy']]

        # incluir puntajes
        player1['score']+=result[0]
        player2['score']+=result[1]

# IMPRIMIR LOS RESULTADOS DE CADA PAR EN CADA RONDA :)
        print(f"{player1['name']} ({player1['strategy']}) vs {player2['name']} ({player2['strategy']}) -> "
              f"Scores: {player1['name']} {player1['score']}, {player2['name']} {player2['score']}")

    print("\nCurrent standings:")
    for player in society:
        print(f"{player['name']} -> Score: {player['score']}")
    print("=" * 50)  # Separador entre rondas
```

```python
# ¿Cómo queda?
society
```

```python
import pandas as pd
```

```python
# organizamos en un data frame
socialResults=pd.DataFrame(society)
socialResults
```

```python
# ¿Quien gana?
winnerScore=socialResults.score.max()

socialResults[socialResults.score==winnerScore]
```

## Ejercicio 2: Aplicación para el juego "Piedra, Papel, Tijera, Lagarto, Spock"

<img src="https://i.imgur.com/DE5mjs4.jpg"/>

**Qué vence a qué**:
   - **Piedra (Rock)**:
     - Aplasta **Tijeras (Scissors)**.
     - Aplasta **Lagarto (Lizard)**.
   - **Papel (Paper)**:
     - Cubre **Piedra (Rock)**.
     - Refuta **Spock**.
   - **Tijera (Scissors)**:
     - Corta **Papel (Paper)**.
     - Decapita **Lagarto (Lizard)**.
   - **Lagarto (Lizard)**:
     - Envenena **Spock**.
     - Come **Papel (Paper)**.
   - **Spock**:
     - Rompe **Tijeras (Scissors)**.
     - Vaporiza **Piedra (Rock)**.

```python
# Estrategias disponibles
strategies2 = ['Rock', 'Paper', 'Scissors', 'Lizard', 'Spock']
```

```python
# Tabla de todos los resultados posibles (payoff)
payoff = {
    ('Rock', 'Scissors'): (1, 0), ('Scissors', 'Rock'): (0, 1),
    ('Rock', 'Lizard'): (1, 0), ('Lizard', 'Rock'): (0, 1),
    ('Paper', 'Rock'): (1, 0), ('Rock', 'Paper'): (0, 1),
    ('Paper', 'Spock'): (1, 0), ('Spock', 'Paper'): (0, 1),
    ('Scissors', 'Paper'): (1, 0), ('Paper', 'Scissors'): (0, 1),
    ('Scissors', 'Lizard'): (1, 0), ('Lizard', 'Scissors'): (0, 1),
    ('Lizard', 'Spock'): (1, 0), ('Spock', 'Lizard'): (0, 1),
    ('Spock', 'Scissors'): (1, 0), ('Scissors', 'Spock'): (0, 1),
    ('Rock', 'Rock'): (0, 0), ('Paper', 'Paper'): (0, 0),
    ('Scissors', 'Scissors'): (0, 0), ('Lizard', 'Lizard'): (0, 0),
    ('Spock', 'Spock'): (0, 0),
}
```

```python
# Inicialización de los jugadores
names = ['rosite', 'aleli', 'ed', 'charles', 'nico', 
         'elenita', 'dani', 'liz', 'tomi', 'jose']
society2 = [{'name': n, 'score': 0, 'strategy': None} for n in names]
```

```python
from random import choice

#simplest strategy: choose randomly
choice(strategies2);
```

```python
# Simulación de varias rondas
for aRound in range(10): 
    print(f"Round {aRound + 1}")
    for player1, player2 in itertools.combinations(society2, 2):  # Enfrentamientos en pares
        # Cada jugador elige una estrategia
        player1['strategy'] = choice(strategies2)
        player2['strategy'] = choice(strategies2)
        
        # Calcular el resultado
        result = payoff.get((player1['strategy'], player2['strategy']), (0, 0))  # Empate por defecto
        player1['score'] += result[0]
        player2['score'] += result[1]
        
        # Mostrar resultados en vivo
        print(f"{player1['name']} ({player1['strategy']}) vs {player2['name']} ({player2['strategy']}) -> "
              f"Scores: {player1['name']} {player1['score']}, {player2['name']} {player2['score']}")
    print("=" * 50)
```

```python
# organizamos en un data frame
socialResults2=pd.DataFrame(society2)
socialResults2
```

```python
# ¿Quien gana?
winnerScore2=socialResults2.score.max()

socialResults2[socialResults2.score==winnerScore2]
```
